```python
# tokenization and lemmatization 
lemmatizer= WordNetLemmatizer()
```

```python
# pokupÄkamo besede s podobnim korenom, pomenom skupaj
# run, runs, running -> run
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('RB'):
        return wordnet.ADV
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    else:
        return wordnet.NOUN
```

```python
def tokenize_lematize(tekst):
    tokens = word_tokenize(tekst.lower())  # vse besede piÅ¡emo z malo zaÄetnico
    tagged = pos_tag(tokens)
    return [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in tagged
        if word.isalpha()    # znebimo se loÄil, st,...
    ]
```

```python
# CountVectorizer odstrani 'stopwords' in ustvari nenegativno matriko, na (i, j)-tem mestu
# imamo pojavitev besede i v j-tem dokumentu (glej zapiske na tablici)


# vzamemo 49/50 knjig, eno bomo potem poskusali uvrstiti med Å¾anre
filepaths = glob.glob(r'C:\Users\mokro\Desktop\diploma\dipl_data\knjige_opisi\*.txt')
# min_df=2, max_df=0.9 odstranita redke in pogoste besede, to uniÄi celoten rezultat
vectorizer= CountVectorizer(stop_words='english', 
                            tokenizer= tokenize_lematize,
                            input = 'filename', 
                            encoding='latin-1', 
                            min_df=2, 
                            max_df=0.9)
```

```python
X = vectorizer.fit_transform(filepaths) 
```

```python
from sklearn.decomposition import NMF
```

```python
# test
k=4
nmf = NMF(
    n_components=k,
    init = 'nndsvd',
    random_state=42,
    max_iter=500
)

W = nmf.fit_transform(X)
H = nmf.components_

#print(W.shape)
#print(H.shape)
#print(f'matrika W:',W)
#print(f'matrika W:',H)
```

```python
k_values = []
errors = []

for k in range(2, 7):
    nmf = NMF(
        n_components=k,
        init='nndsvd',
        random_state=42
    )

    W = nmf.fit_transform(X)
    H = nmf.components_



    feature_names = vectorizer.get_feature_names_out()
    print('k je', k)
    for topic_idx, topic in enumerate(H):
        top_words = [feature_names[i] for i in topic.argsort()[-10:]]
        print(f"Tema {topic_idx}: {' '.join(top_words)}")

    error = nmf.reconstruction_err_

    print('k=', k, 'reconstruction error =', error)

    k_values.append(k)
    errors.append(error)
    

# naredi test ki ti bo povedal kateri k je najboljÅ¡i reconstruction error
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(k_values, errors, marker='o')
plt.xlabel("Number of Topics (k)")
plt.ylabel("Reconstruction Error")
plt.title("NMF Reconstruction Error vs Number of Topics")

plt.grid(True)

plt.show()
```
